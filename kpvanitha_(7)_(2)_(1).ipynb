Similarity Matrix Preparation


```python
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
#Cosine Similarity to compute the similarity between a pair of sentences.
from sklearn.metrics.pairwise import cosine_similarity
```

Create a dictionary of words and their corresponding index:



```python
dictionary = {}
index = 0

for words in lemtext:
    for word in words:
        if word not in dictionary:
            dictionary[word] = index
            index += 1
```

```python
# create a numpy array to hold the similarity scores
sentences = sent_tokenize(stopword_removed_text)
num_sentences = len(sentences)
similarity_matrix = np.zeros((num_sentences, num_sentences))

```

```python

# compute the similarity scores using cosine similarity
for i in range(num_sentences):
    for j in range(num_sentences):
        if i == j:
            continue
        vector1 = np.zeros(len(dictionary))
        vector2 = np.zeros(len(dictionary))
        words1 = lemtext[i]
        words2 = lemtext[j]
        for word in set(words1 + words2):
            if word in dictionary:
                vector1[dictionary[word]] = words1.count(word)
                vector2[dictionary[word]] = words2.count(word)
        similarity_matrix[i][j] = cosine_similarity([vector1], [vector2])[0, 0]
```

```python
import pandas as pd
similarity_df = pd.DataFrame(similarity_matrix)
print(similarity_df)

```

Applying PageRank Algorithm
Before proceeding further, let’s convert the similarity matrix sim_mat into a graph. The nodes of this graph will represent the sentences and the edges will represent the similarity scores between the sentences. On this graph, we will apply the PageRank algorithm to arrive at the sentence rankings.

```python
import networkx as nx

nx_graph = nx.from_numpy_array(similarity_matrix)
scores = nx.pagerank(nx_graph)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Your code to create nx_graph and scores
nx_graph = nx.from_numpy_array(similarity_matrix)
scores = nx.pagerank(nx_graph)

# Create a list of scores
score_list = list(scores.values())

# Create a histogram with 10 bins
plt.hist(score_list, bins=10)


# Add labels and title to the histogram
plt.xlabel('PageRank Scores')
plt.ylabel('Frequency')
plt.title('Histogram of PageRank Scores')
# Show the histogram
plt.show()
```
